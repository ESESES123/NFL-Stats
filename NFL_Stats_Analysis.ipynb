# Predicting Game Outcomes
Mergeddataset combines the QB stats with the defensive stats

```python
Mergeddataset=Gamedataset.merge(Defensedataset,on=["Team","Year"]).merge(QBdataset,on=["Year","Name"],suffixes=["_Defense","_QB"])
Mergeddataset["Outcome"]=Mergeddataset["Outcome"].replace({"W":1,"L":0,"T":0})
```

We considered categories of NFL stats to predict the chance the QB would win. 0.89 means that we can correctly predict the winner 89% of the time.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
#lr=LogisticRegression()
lr=RandomForestClassifier()
x=Mergeddataset[["Total Tackles","Solo Tackles","Assisted Tackles","Sacks_Defense","Safties","Ints_Defense","Ints for TDs","Int Yards","Games Played_QB","Passing Yards","Passer Rating","Completion Percentage","Sacks_QB","Ints_QB","TD Passes","Fumbles Lost"]]
y=Mergeddataset.Outcome
lr.fit(x,y)
lr.score(x,y)
```

By using the code below we could input a QB, a defense, and a year and predict the chances of a QB winning against a team in a given year.

```python
def PredictWinner(Team, Name, Year):
  Comparisondata=Defensedataset[(Defensedataset.Team==Team)&(Defensedataset.Year==Year)].merge(QBdataset[(QBdataset.Name==Name)&(QBdataset.Year==Year)],on=["Year"],suffixes=["_Defense","_QB"])
  return lr.predict_proba(Comparisondata[["Total Tackles","Solo Tackles","Assisted Tackles","Sacks_Defense","Safties","Ints_Defense","Ints for TDs","Int Yards","Games Played_QB","Passing Yards","Passer Rating","Completion Percentage","Sacks_QB","Ints_QB","TD Passes","Fumbles Lost"]])[0,1]
```

```python
print("The chance of Peyton Manning winning against the Patriots is",PredictWinner("New England Patriots","Manning, Peyton",2014))
print("The chance of Tom Brady winning against the Oakland Raiders in 2014 is",PredictWinner("Oakland Raiders","Brady, Tom",2014))
```

```python
probabilities=[]
for i in [2005,2006,2007,2008,2009,2010,2011,2012,2013]:
  probabilities.append (PredictWinner("Oakland Raiders","Brady, Tom",i))
fig=px.line(y=probabilities,x=[2005,2006,2007,2008,2009,2010,2011,2012,2013],title="Probability of Tom Brady Winning vs the Raiders")
fig.update_layout(xaxis_title='Year', yaxis_title='Win Probability')
```

```python
probabilities=[]
for i in ["Oakland Raiders","Green Bay Packers","Kansas City Chiefs","Denver Broncos","San Francisco 49ers"]:
  probabilities.append (PredictWinner(i,"Brady, Tom",2014))
fig=px.bar(y=probabilities,x=["Oakland Raiders","Green Bay Packers","Kansas City Chiefs","Denver Broncos","San Francisco 49ers"],title="Probablity of Tom Brady Winning against certain teams in 2014")
fig.update_layout(xaxis_title='Team', yaxis_title='Win Probability')
```

```python
probabilities=[]
for i in ["Manning, Peyton","Rodgers, Aaron","Wilson, Russell","Palmer, Carson","Bortles, Blake"]:
  probabilities.append (PredictWinner("New England Patriots",i,2015))
fig=px.bar(y=probabilities,x=["Manning, Peyton","Rodgers, Aaron","Wilson, Russell","Palmer, Carson","Bortles, Blake"],title="Probability of QBs winning against the Patriots in 2015")
fig.update_layout(xaxis_title='Quarterback', yaxis_title='Win Probability')
```

```python
def RankTeams(Year):
  import numpy as np
  probabilities=[]
  for i in QBdataset[(QBdataset.Year==Year)&(QBdataset["Games Played"]>0.7)].Name:
    QBProbabilities=[]
    for j in Defensedataset[Defensedataset.Year==Year].Team:
      QBProbabilities.append (PredictWinner(j,i,Year))
    probabilities.append(np.mean(QBProbabilities))
  figure=px.bar(y=probabilities,x=QBdataset[(QBdataset.Year==Year)&(QBdataset["Games Played"]>0.7)].Name,title="Win probability of QBs in 1990")
  figure.update_layout(xaxis={"categoryorder":"total descending"})
  figure.update_layout(xaxis_title='Quarterback', yaxis_title='Win Probability')
  figure.show()
  import numpy as np
  probabilities=[]
  for j in Defensedataset[Defensedataset.Year==Year].Team:
    QBProbabilities=[]
    for i in QBdataset[(QBdataset.Year==Year)&(QBdataset["Games Played"]>0.7)].Name:
      QBProbabilities.append (PredictWinner(j,i,Year))
    probabilities.append(np.mean(QBProbabilities))
  figure=px.bar(y=probabilities,x=Defensedataset[(Defensedataset.Year==Year)].Team,title="Win probability of Defenses in 1990")
  figure.update_layout(xaxis={"categoryorder":"total descending"})
  figure.update_layout(xaxis_title='Defense', yaxis_title='Win Probability')
  figure.show()
```
